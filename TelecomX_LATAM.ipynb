RECONTEO DE EVASION POR VARIABLES CATEGORICAS

```python

# ANÁLISIS: EVASIÓN POR VARIABLES (CON DESCRIPCIÓN)

import seaborn as sns
import matplotlib.pyplot as plt

# 1. Definimos las variables clave
variables = ['genero', 'tipo_contrato', 'tipo_internet']

for columna in variables:
    plt.figure(figsize=(10, 5))

    # Creamos el gráfico de barras comparativo
    ax = sns.countplot(data=df, x=columna, hue='evasion', palette=['#3498DB', '#E67E22'])

    # Agregamos etiquetas de datos sobre las barras
    for container in ax.containers:
        ax.bar_label(container)

    plt.title(f'Impacto de "{columna}" en la Evasión de Clientes', fontsize=14, fontweight='bold')
    plt.legend(title='Evasión', labels=['Se quedaron (0)', 'Se fueron (1)'])
    sns.despine()
    plt.show()

    # --- DESCRIPCIÓN DEL ANÁLISIS ---
    print(f"¿QUÉ NOS DICE ESTA GRÁFICA SOBRE '{columna.upper()}'?")

    if columna == 'genero':
        print("- Observamos que tanto hombres como mujeres tienen un comportamiento de evasión muy similar.")
        print("- Conclusión: El género NO es un factor determinante para que un cliente abandone la empresa.")

    elif columna == 'tipo_contrato':
        print("- ¡ALERTA! Los clientes con contrato 'Mensual' tienen la mayor tasa de abandono.")
        print("- Los contratos Anuales y Bienales muestran una lealtad mucho más alta.")
        print("- Conclusión: El compromiso a largo plazo reduce drásticamente la evasión.")

    elif columna == 'tipo_internet':
        print("- Nota cómo los usuarios de 'Fibra óptica' suelen tener una barra de evasión proporcionalmente más alta que DSL.")
        print("- Conclusión: Podría haber un problema de precio o estabilidad en el servicio de fibra que molesta a los clientes.")

    print("\n" + "="*80 + "\n")
```

# 📊 Informe Final: Análisis de Evasión de Clientes – Telecom X

## 1. Introducción

El objetivo de este proyecto es entender por qué algunos clientes dejan de usar los servicios de **Telecom X**.  
A este fenómeno se le llama **Churn** o **evasión de clientes**.

En el sector de telecomunicaciones, perder clientes es un problema importante porque **conseguir nuevos clientes cuesta mucho más que mantener a los actuales**.

Por eso, en este análisis buscamos:

- Identificar **qué tipo de clientes tienen mayor riesgo de irse**
- Detectar **las razones principales por las que cancelan el servicio**
- Encontrar **oportunidades para mejorar la retención de clientes**

---

## 2. Limpieza y Preparación de Datos

Antes de analizar la información, fue necesario preparar y ordenar los datos.

Se realizaron los siguientes pasos:

### Extracción de datos
Los datos estaban en formato **JSON** y fueron convertidos a un **DataFrame de Pandas** para poder analizarlos.

### Eliminación de inconsistencias
Se eliminaron registros que tenían valores vacíos en la variable principal **evasión**, ya que no permitían identificar si el cliente se fue o no.

### Corrección de datos financieros
En la columna **cargo_total** se encontraron espacios en blanco.  
Estos valores fueron reemplazados por **0** y luego convertidos a formato numérico para permitir cálculos.

### Estandarización
Para facilitar el análisis:

- Los nombres de las columnas se tradujeron al **español**
- Las respuestas **Sí / No** se transformaron en valores **1 y 0**
- Esto permite trabajar mejor con modelos y análisis estadísticos

---

## 3. Análisis Exploratorio de Datos (EDA)

Después de visualizar y explorar los datos, se identificaron varios patrones importantes.

### 📉 Tasa de evasión

La tasa de evasión es **26.5%**.

Esto significa que **aproximadamente 1 de cada 4 clientes deja la empresa**.

---

### 📄 Tipo de contrato

El tipo de contrato influye mucho en la permanencia del cliente.

- Los clientes con contrato **Mes a mes** son los que **más cancelan el servicio**.
- Los contratos **anuales o de dos años** muestran **mayor fidelidad**.

Esto indica que **los contratos largos ayudan a retener clientes**.

---

### 🌐 Servicio de internet

Se observó una **mayor evasión entre clientes que usan Fibra Óptica**.

Esto podría indicar:

- Problemas de **precio**
- Problemas de **calidad o estabilidad del servicio**
- **Mejores ofertas de la competencia**

---

### 💰 Facturación mensual

Los clientes que cancelan el servicio suelen tener **cargos mensuales más altos**.

Los gráficos de correlación y los **boxplots** muestran que **los precios elevados aumentan la probabilidad de evasión**.

---

## 4. Conclusiones e Insights

### ⏱ La barrera de los 6 meses

Muchas cancelaciones ocurren **durante los primeros meses del servicio**.

Si un cliente **permanece más de un año**, es mucho menos probable que abandone la empresa.

---

### 💸 Sensibilidad al precio

El **costo mensual** es uno de los factores más importantes en la decisión de cancelar el servicio.

Cuando el precio es alto, algunos clientes **no perciben suficiente valor**.

---

### 👥 El género no influye

No se encontraron diferencias significativas entre **hombres y mujeres** en la tasa de evasión.

Esto significa que **el género no es un factor relevante en la cancelación del servicio**.

---
